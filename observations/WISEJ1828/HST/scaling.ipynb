```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
import sys
import pandas as pd
from pathlib import Path
os.environ['pRT_input_data_path'] = os.path.join('/home/mvasist/pRT/input_data_paper/input_data')
from petitRADTRANS import Radtrans
from petitRADTRANS import nat_cst as nc
```

```python
home = os.environ.get('HOME')
path = Path(home)
```

```python
obsHST_file = path / 'WISEJ1828' / 'WISEJ1828' / '6_HST' / 'WISE1828.fl.txt'
obs_hst_pd = pd.read_csv(obsHST_file, delim_whitespace= True, header=1)
obs_wlen_hst = np.array(obs_hst_pd.iloc[:,0])
obs_hst = np.array(obs_hst_pd.iloc[:, 1]*1e30 * (obs_wlen_hst*1e-4)**2/nc.c)
sigmaH = np.array((obs_hst_pd.iloc[:, 2]*1e30 * (obs_wlen_hst*1e-4)**2/nc.c))
```

```python
spec_hst = np.stack((obs_wlen_hst, obs_hst, sigmaH), axis=1)
df = pd.DataFrame(spec_hst) #convert to a dataframe
df.to_csv(path / 'SBI4exoplanets'/ 'observations'/ "WISEJ1828" / "HST"/"spectrum.csv",index=False, float_format="%.16f") #save to file
```
