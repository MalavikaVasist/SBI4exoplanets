```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
```

```python
channels = ['1A', '1B', '1C', '2A', '2B', '2C', '3A', '3B', '3C']

spec= pd.DataFrame()
# spectrum={}
# c = ['black', 'green', 'blue', 'pink', 'purple', 'red', 'yellow', 'orange', 'brown']
# length=0

for channel in channels:
    path = '/home/mvasist/WISEJ1828/WISEJ1828/3/separated_WISE1828_spectrum_200223_send/' + channel + '.dat'
    print(path)
    spec_per_channel = pd.read_csv(path,delim_whitespace= True, header=0)
    spec_per_channel = spec_per_channel.rename({'#': 'wavelength [um]', 'central': 'flux [Jy]', 'wavelength[um]': 'error [Jy]', 'flux[Jy]': 'wavelength bin width [um]'}, axis = 'columns')
    spec_per_channel.drop(columns=spec_per_channel.columns[-4:], axis=1,  inplace=True)
    spec = spec.append(spec_per_channel)
#     spectrum[path.split('send/')[1].split('.')[0]]= spec_per_channel
    
spec
```

```python
newDf = pd.DataFrame([])
wold = spec['wavelength [um]'].iloc[0]
for w in spec['wavelength [um]'].iloc[:]:
    if w > wold: 
        a = spec[spec['wavelength [um]'] == w]
        newDf = newDf.append(a.mean(axis=0),ignore_index=True)
        wold = w
```

```python
newDf.to_csv('spectrum.csv', index=False)
```
