## Convolving to 130 

```python
scratch = os.path.join('/scratch/users/mvasist') #os.environ['SCRATCH'] = 
datapath = Path(scratch) / 'JWST/simulations/sim_15NH3HCNTiOVOXH2OcLPbasefsedkzzslnbfacM_unregPT/cloudfree/HST/simulation'
trainset = H5Dataset(datapath / 'train.h5')
```

```python
def wlenbins(w1):
    wlen_bins = np.zeros_like(w1)
    wlen_bins[:-1] = np.diff(w1)
    wlen_bins[-1] = wlen_bins[-2]
    return wlen_bins

pRT_object = Radtrans(['CH4_hargreaves',
                        'H2O_Exomol',
                        'SO2'],
                     continuum_opacities=['H2-H2', 'H2-He'],
                     rayleigh_species=['He', 'H2'],
                     wlen_bords_micron=[0.98, 2.2])  #4.8, 18.1
wlen = nc.c/pRT_object.freq/1e-4
```

```python
#1.1 - 1.69um
for i, (theta, x) in enumerate(trainset):
    xx = Data.convolve(wlen[115:552], x[115:552], 130) #252, 753
    break
wlen_bins = wlenbins(obs_wlen_hst)
flux_rebinned = torch.from_numpy(rebin_give_width(wlen[115:552], xx, obs_wlen_hst, wlen_bins))
```
