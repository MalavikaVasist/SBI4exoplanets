```python
from astropy.io import fits
from astropy.table import Table
```

```python
path = Path('reprocessed231123/fits')
```

```python
l = 'Level3_'
x = '_x1d'
tbpd = pd.DataFrame()
a = ['1A', '1B', '1C', '2A', '2B', '2C', '3A', '3B', '3C']
k = 0 

for i, ch in enumerate(['ch1', 'ch2', 'ch3']):
    for j, length in enumerate(['short', 'medium', 'long']):
        pp = l + ch + '-' + length + x 
        hdul = fits.open(path / (pp + '.fits'))
        asn_table = Table(hdul[1].data)
        ppdd = pd.DataFrame(np.array(asn_table))
        df = pd.DataFrame({'Band': [str(a[int(k)])]*len(asn_table)})
        ppdd = pd.concat([ppdd, df], axis=1)
        tbpd = pd.concat([tbpd, ppdd], ignore_index=True)
        k+=1
        print(k)
        hdul.close()
```
