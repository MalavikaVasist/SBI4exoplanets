```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
import sys
import pandas as pd
from pathlib import Path

home = os.environ.get('HOME')
path = Path(home)
```

```python
obsMIRI_file = path / 'SBI4exoplanets'/ 'observations'/ 'WISEJ1738'/ 'MIRI' / 'Helena_rebinning_all_ch_together' / 'MIRI1000_1278.dat'
obs_miri_pd = pd.read_csv(obsMIRI_file)
helena_spectrum = obs_miri_pd.values
# helena_spec = torch.Tensor([np.array(helena_spectrum[i][0].split(), dtype=float) for i, hs in enumerate(helena_spectrum)])
# Convert list of NumPy arrays into a single NumPy array
helena_spec_np = np.array([np.array(hs[0].split(), dtype=float) for hs in helena_spectrum])
# Convert NumPy array to PyTorch tensor
helena_spec = torch.tensor(helena_spec_np, dtype=torch.float)

obs_wlen_miri  = np.array(helena_spec[:,0])
obs_miri = np.array(helena_spec[:,2])
sigmaM = np.array(helena_spec[:,3])

obsMIRI_file_ = path / 'SBI4exoplanets'/ 'observations'/ 'WISEJ1738'/ 'MIRI' / 'spectrum.csv'
obs_miri_pd_ = pd.read_csv(obsMIRI_file_)
obs_wlen_miri_  = np.array(obs_miri_pd_.iloc[:, 0])
obs_miri_ = np.array(obs_miri_pd_.iloc[:, 1])
sigmaM_ = np.array(obs_miri_pd_.iloc[:, 2])


before_col =  [obs_miri_[0]] # Column to append before
after_col =  [obs_miri_[-1]]   # Column to append after
result_col = np.hstack((before_col, obs_miri, after_col))

before_col_M =  [sigmaM_[0]] # Column to append before
after_col_M =  [sigmaM_[-1]]   # Column to append after
result_col_M = np.hstack((before_col_M, sigmaM, after_col_M))


before_col_w =  [obs_wlen_miri_[0]] # Column to append before
after_col_w =  [obs_wlen_miri_[-1]]   # Column to append after
result_col_w = np.hstack((before_col_w, obs_wlen_miri, after_col_w))
```

```python
spec = np.stack((result_col_w, result_col, result_col_M), axis= 1)
```

```python
df = pd.DataFrame(spec) #convert to a dataframe
df.to_csv("spectrum.csv",index=False, float_format="%.16f") #save to file
```
