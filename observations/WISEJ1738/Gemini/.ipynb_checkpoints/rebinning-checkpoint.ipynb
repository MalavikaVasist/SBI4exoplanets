```python
import numpy as np
import pandas as pd
from pathlib import Path
import torch
import matplotlib.pyplot as plt
import os
os.environ['pRT_input_data_path'] = os.path.join('/home/mvasist/pRT/input_data_paper/input_data')

from petitRADTRANS.retrieval.rebin_give_width import rebin_give_width
from petitRADTRANS.retrieval.data import Data
from petitRADTRANS import Radtrans
from petitRADTRANS import nat_cst as nc

from lampe.data import H5Dataset

scratch = os.environ.get('SCRATCH', '')
```

```python
def wlenbins(w1):
    wlen_bins = np.zeros_like(w1)
    wlen_bins[:-1] = np.diff(w1)
    wlen_bins[-1] = wlen_bins[-2]
    return wlen_bins

def simulating_wavelengths(a,b):
    
    pRT_object = Radtrans(['CH4_hargreaves',
                        'H2O_Exomol',
                        'SO2'],
                     continuum_opacities=['H2-H2', 'H2-He'],
                     rayleigh_species=['He', 'H2'],
                     wlen_bords_micron=[a,b])  
    
    wlen = nc.c/pRT_object.freq/1e-4
    
    return wlen

def flux_rebin(w , fl, fl_err):
    mirisim_wlen = simulating_wavelengths(w[0]+0.01, w[-1]-0.01)
    wlen_bins = wlenbins(mirisim_wlen)
    flux_rebinned = torch.from_numpy(rebin_give_width(w, fl, mirisim_wlen, wlen_bins))
    flux_rebinned_err = torch.from_numpy(rebin_give_width(w, fl_err, mirisim_wlen, wlen_bins))
    return torch.stack((torch.from_numpy(mirisim_wlen), flux_rebinned, flux_rebinned_err), dim=1)
```

```python
path = Path('.') / 'WISEJ1738_Gemini.txt'

spec= pd.DataFrame()
spec = pd.read_csv(path, header=0, delimiter = '\t')
specc = spec.astype(np.float64)
w = specc.iloc[:,0].values
```

## Flambda --> fnu --> rebin 

```python
fl = specc.iloc[:,1].values * 1e7 * 1e23 * (specc.iloc[:,0].values*1e-4)**2. / nc.c  
fl_err = specc.iloc[:,2].values * 1e7 * 1e23 * (specc.iloc[:,0].values*1e-4)**2. / nc.c
```

```python
sp = torch.stack((torch.from_numpy(w), torch.from_numpy(fl), torch.from_numpy(fl_err)))
sp = torch.permute(sp, (1,0))
```

```python
sp.size()
```

```python
def gap_finder_(spec):
    plt.scatter(spec[:,0], spec[:,0])
    plt.show()
    d = np.diff(spec[:,0])
    plt.plot(spec[:-1,0], d)
    plt.show()
    limit = np.where(d>0.05)[0]
    limit = limit +1
    limit = np.insert(limit, 0, 0)
    limit = np.insert(limit, 4, len(spec[:,0]))
    return limit
limit = gap_finder_(sp)
```

```python
i = 0
specrb = [flux_rebin(w[limit[i]:limit[i+1]], fl[limit[i]:limit[i+1]], fl_err[limit[i]:limit[i+1]]) for i in range(4)]
```

```python
specrb = torch.vstack(specrb)
```

```python
plt.plot(specrb[:,0], specrb[:,1]*1e5 * (7.34/9.9)**2)
plt.ylabel('flux Jy')
plt.title('as the NN sees it')
plt.legend()
plt.show()
```
